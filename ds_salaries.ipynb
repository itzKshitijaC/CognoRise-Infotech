```python
# Job Title with the highest salary in usd

plt.figure(figsize=(12,6))
top10JobTitlesHighestSalary = sns.barplot(df.groupby(['job_title'])['salary_in_usd'].sum().iloc[:10].sort_values(ascending=False), palette="summer")
for bars in top10JobTitlesHighestSalary.containers:
    top10JobTitlesHighestSalary.bar_label(bars)
plt.title('Job Title with the highest salary in usd')
plt.xlabel("Job Title")
plt.ylabel("Salary in usd")
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```

```python
# Job Title with the highest salary

plt.figure(figsize=(12,6))
top10JobTitlesHighestSalary = sns.barplot(df.groupby(['job_title'])['salary'].sum().iloc[:10].sort_values(ascending=False), palette="Set3")
for bars in top10JobTitlesHighestSalary.containers:
    top10JobTitlesHighestSalary.bar_label(bars)
plt.title('Job Title with the highest salary')
plt.xlabel("Job Title")
plt.ylabel("Salary in usd")
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```

```python
# Job Title with the minimum salary

plt.figure(figsize=(12,6))
top10JobTitlesLowestSalary = sns.barplot(df.groupby(['job_title'])['salary'].sum().iloc[-10:].sort_values(ascending=False), palette="winter")
for bars in top10JobTitlesLowestSalary.containers:
    top10JobTitlesLowestSalary.bar_label(bars)
plt.title('Job Title with the lowest salary')
plt.xlabel("Job Title")
plt.ylabel("Salary in usd")
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```

```python
# Salary with experience level
salarybyexplevel = df.groupby(['experience_level'])['salary'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.pie(salarybyexplevel, labels=salarybyexplevel.index, autopct='%.0f%%', startangle=140)
plt.title("Salary by Experience level", fontsize=14)
plt.axis('equal')  
plt.show()
```

```python
# Salary with Employment Type
plt.figure(figsize=(12,6))
salarybyEmploymentType = sns.barplot(df.groupby(['employment_type'])['salary_in_usd'].sum().iloc[-10:].sort_values(ascending=False), palette="winter")
for bars in salarybyEmploymentType.containers:
    salarybyEmploymentType.bar_label(bars)
plt.title('Salary in usd by Employment Type')
plt.xlabel("Job Title")
plt.ylabel("Salary in usd")
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```

```python
# Salary by Company Size
salarybyCompanySize = df.groupby(['company_size'])['salary'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,4))
plt.pie(salarybyCompanySize, labels=salarybyCompanySize.index, autopct='%.0f%%', startangle=140)
plt.title("Salary by Company Size", fontsize=14)
plt.axis('equal')  
plt.show()
```

```python
# Salary in usd by Company Size
salarybyCompanySize = df.groupby(['company_size'])['salary_in_usd'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,4))
plt.pie(salarybyCompanySize, labels=salarybyCompanySize.index, autopct='%.0f%%', startangle=140)
plt.title("Salary by Company Size", fontsize=14)
plt.axis('equal')  
plt.show()
```

Analyze the relationship between company size, employment type, and salary. Do large companies offer higher salaries than medium or small companies? How does the employment type (FT) influence this relationship?

```python

ft_data = df[df['employment_type'] == 'FT']

plt.figure(figsize=(12, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=ft_data, palette='magma')

plt.title('Salary Distribution by Company Size for Full-Time Employment')
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.show()
```

Compare the salaries of employees based on their residence. Are there disparities in salaries for employees residing in different countries but working in the same job roles?

```python
average_salary = df.groupby(['job_title', 'employee_residence'])['salary_in_usd'].mean().reset_index()

plt.figure(figsize=(16, 8))
sns.barplot(x='employee_residence', y='salary_in_usd', hue='job_title', data=average_salary, palette='viridis')

plt.title('Average Salary by Job Title and Employee Residence')
plt.xlabel('Employee Residence')
plt.ylabel('Average Salary in USD')
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

How does the company location affect the salary in USD? Are there specific regions where data science professionals are paid significantly higher or lower?

```python
average_salary_by_location = df.groupby('company_location')['salary_in_usd'].mean().reset_index()
plt.figure(figsize=(16, 8))
sns.barplot(x='company_location', y='salary_in_usd', data=average_salary_by_location, palette='coolwarm')

plt.title('Average Salary by Company Location')
plt.xlabel('Company Location')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45)
plt.show()
```

Is there a correlation between the remote ratio and the salary in USD? Does working remotely (50% or 100%) have a significant impact on salaries compared to on-site work?

```python
correlation = df['remote_ratio'].corr(df['salary_in_usd'])
print(f"Correlation between remote ratio and salary in USD: {correlation}")
plt.figure(figsize=(12, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=data, palette='Set3')

plt.title('Salary Distribution by Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary in USD')
plt.show()
```

How have salaries in USD for data science professionals changed over the years in the dataset? Are there observable trends or patterns in salary growth or decline for different job titles or experience levels?

```python

average_salary_by_year = df.groupby(['work_year', 'job_title', 'experience_level'])['salary_in_usd'].mean().reset_index()
average_salary_by_year['job_exp'] = average_salary_by_year['job_title'] + ' (' + average_salary_by_year['experience_level'] + ')'
plt.figure(figsize=(16, 8))
sns.lineplot(x='work_year', y='salary_in_usd', hue='job_exp', data=average_salary_by_year, markers=True, dashes=False, palette='tab20')
plt.title('Salary Trends Over the Years by Job Title and Experience Level')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.legend(title='Job Title (Experience Level)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(average_salary_by_year['work_year'].unique())
plt.grid(True)
plt.tight_layout()
plt.show()
```
